```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

# Shock Wave and Mach Number Relationship

```python
Shock_Data = {}
for file in ['1.3', '3.17', '5.0']:
    df = pd.read_csv(os.path.join('data', '0.01', file))
    Shock_Data[file] = df
    

for M in Shock_Data.keys():
    plt.scatter(Shock_Data[M]['x'], Shock_Data[M]['y'], label=M)

plt.legend()  
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0.0, 0.5)
plt.ylim(0.0, 0.2)
plt.title('Elliptical Airfoil Detached Bow Shock h = 0.01')
plt.annotate('Low Mach\n Error', xy=(0.3, 0.025), xytext=(0.4, 0.040),arrowprops=dict(facecolor='black', shrink=0.1, linewidth=0.2))
plt.annotate('Low Mach\n Error', xy=(0.37, 0.055), xytext=(0.4, 0.040),arrowprops=dict(facecolor='black', shrink=0.1, linewidth=0.2))
```

## The lower the Mach number, the larger the shock deflection upwards. This idea matches theory for supersonic flow.

# Shockwave and Airfoil Thickness Relationship

```python
Shock_Data = {}
for file in ['0.005', '0.0072', '0.01']:
    df = pd.read_csv(os.path.join('data', file, '5.0'))
    Shock_Data[file] = df
    

for M in Shock_Data.keys():
    plt.scatter(Shock_Data[M]['x'], Shock_Data[M]['y'], label=M)
    
plt.legend()  
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0.0, 0.5)
plt.ylim(0.0, 0.2)
plt.title('Elliptical Airfoil Detached Bow Shock')
```

## The larger the height of the ellipse, the larger the shock deflection upwards. This idea matches theory for supersonic flow.
